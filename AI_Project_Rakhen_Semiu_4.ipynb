# Implementing PSO in Greedy Epsilon for Multi-Armed Bandit Problem

Importing the required libraries 

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

Assigning the number of the slot machines which is also used as the number of particles in the PSO

```python
number_of_slot_machines = int(input("The number of the slot machine: "))  
```

Creates an array of the size of slot machine with Jackpot Probability based on the range

```python
np.random.seed(100)
JPs =  np.random.uniform(0,1, number_of_slot_machines)
```

Formalization of the Agent Playing at the each machine

```python
def play_machine(slot_machine):
  x = np.random.uniform(0, 1)
  if (x <= JPs[slot_machine]):
    return(10)
  else:
    return(-1)
```

Assigning the number of iterations

```python
n_iterations = int(input("The number of iteration: "))
```

Defining the target_error as the maximum value of JPs. This is used in the PSO search oprtimization

```python
target_error = np.amax(JPs)
```

Search Space Optimization Function using Particle Swarm Optimization to get "gbest_fitness_value" which is later used in the Reinforement Learning code

```python
#function that models the problem
def fitness_function(position):
    return position[0]**2 + position[1]**2 + 1
    return(gbest_fitness_value)

#Some variables (weight, coeeficient of the accelerations, and maximum target) to calculate the velocity
W = 0.5
c1 = 0.5
c2 = 0.9
target = 1
#target_error = 0.8

#Definitions of other parameters used in the implementations of the PSO:
#a. PSO number of particles is assigned number_of_slot_machines of the Multi-armed bandit.
#b. The number of iterations is earlier assigned through an input function.
#c. The target_error is defined as the maximum number of the JPs.

#target_error = float(input("Inform the target error: "))

particle_position_vector = np.array([np.array([(-1) ** (bool(random.getrandbits(1))) * random.random()*50, (-1)**(bool(random.getrandbits(1))) * random.random()*50]) for _ in range(number_of_slot_machines)])

pbest_position = particle_position_vector
pbest_fitness_value = np.array([float('inf') for _ in range(number_of_slot_machines)])
gbest_fitness_value = float('inf')
gbest_position = np.array([float('inf'), float('inf')])

velocity_vector = ([np.array([0, 0]) for _ in range(number_of_slot_machines)])
initial_iteration = 0
while initial_iteration < n_iterations:
    for i in range(number_of_slot_machines):
        fitness_cadidate = fitness_function(particle_position_vector[i])
        print(fitness_cadidate, ' ', particle_position_vector[i])
        
        if(pbest_fitness_value[i] > fitness_cadidate):
            pbest_fitness_value[i] = fitness_cadidate
            pbest_position[i] = particle_position_vector[i]

        if(gbest_fitness_value > fitness_cadidate):
            gbest_fitness_value = fitness_cadidate
            gbest_position = particle_position_vector[i]

    if(abs(gbest_fitness_value - target) < target_error):
        break
    
    for i in range(number_of_slot_machines):
        new_velocity = (W*velocity_vector[i]) + (c1*random.random()) * (pbest_position[i] - particle_position_vector[i]) + (c2*random.random()) * (gbest_position-particle_position_vector[i])
        new_position = new_velocity + particle_position_vector[i]
        particle_position_vector[i] = new_position

    initial_iteration = initial_iteration + 1
print (gbest_fitness_value)
print (gbest_position)
type (gbest_position)
type (gbest_fitness_value)
```

Reinforcement Learning Code


```python
def multi_armed_bandit(arms, iterations):
    total_reward =  []
    optimal_action = []
    estimated_payout_odds = np.zeros(arms)
    count = np.zeros(arms)
    for i in range(0, iterations):
        #The NP Inverse Logarithm gbest_fitness_value to implement Greedy-Epsilon-PSO 
        epsilon = 1/np.log(gbest_fitness_value)
        epsilon_random = np.random.uniform(0, 1)
        if epsilon_random > epsilon:
            # exploit
            action = np.argmax(estimated_payout_odds)
        else:
            # explore
            action = np.random.choice(np.arange(arms))
            
        reward = play_machine(action)
        estimated_payout_odds[action] = estimated_payout_odds[action] + (1/(count[action]+1)) * (reward - estimated_payout_odds[action])
        
        total_reward.append(reward)
        optimal_action.append(action == np.argmax(estimated_payout_odds))
        count[action] += 1
        
    return(estimated_payout_odds, total_reward)
```

```python
print("-------------------------------------------")
print ("Learned Odds with decaying value of epsilon")
print("-------------------------------------------")
learned_payout_odds, reward = multi_armed_bandit(number_of_slot_machines, n_iterations)
plt.bar(np.arange(len(learned_payout_odds)), learned_payout_odds)
plt.show()
print(np.round(learned_payout_odds,2))
print("Reward:", sum(reward))
```
